```python
import pandas as pd
import os
import sys
import urllib.request
import json
import re
```

```python
def clean_html(x):
  #https://predictor-ver1.tistory.com/4
  x = re.sub("\&\w*\;","",x)
  x = re.sub("<.*?>","",x)
  return x
```
