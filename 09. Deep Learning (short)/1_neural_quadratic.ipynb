## 1.Import Required Libraries

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.init as init
from torch.autograd import Variable
# from visdom import Visdom
# viz = Visdom()
```

## 2. Data Generation

```python
num_data = 1000
num_epoch = 50000

noise = init.normal(torch.FloatTensor(num_data,1),std=3)

x = init.uniform(torch.Tensor(num_data,1),-15,15)
y = (x**2) + 3 

y_noise = y + noise
```

## 3. Model & Optimizer

```python
# fully connected model with 4 hidden layer

model = nn.Sequential(
        nn.Linear(1,5),
        nn.ReLU(),
        nn.Linear(5,50),
        nn.ReLU(),
        nn.Linear(50,100),
        nn.ReLU(),
        nn.Linear(100,200),
        nn.ReLU(),
        nn.Linear(200,1)
    )

loss_func = nn.L1Loss()
optimizer = optim.SGD(model.parameters(),lr=0.001)
```

## 4. Train 

```python
loss_arr =[]
label = y_noise

for i in range(num_epoch):
    optimizer.zero_grad()
    output = model(x)

    loss = loss_func(output,label)
    loss.backward() #여기서 W와 b의 미분값을 구합니다.
    optimizer.step() #구해진 미분값을 기준으로 W와 b값을 업데이트 합니다.
    
    if i % 100 ==0:
        print(loss)
    
    #loss_arr.append(loss.cpu().data.numpy()[0])
    loss_arr.append(loss.item())
```
