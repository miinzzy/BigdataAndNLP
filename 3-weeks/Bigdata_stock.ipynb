```python
import numpy as np
import pandas as pd

nc = pd.read_csv('/content/sample_data/price.csv')
nc
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(18,12))

plt.title("Keyword", fontsize = 30)

plt.plot(nc['date'], nc['pattern'], 'g', label='리니지 문양 사건')
plt.plot(nc['date'], nc['NFT'], 'r', label='NC 소프트 NFT')
plt.plot(nc['date'], nc['BAS'], 'b', label='블레이드 앤 소울 2')

plt.xticks(rotation = 45)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(18,15))
plt.xticks(rotation = 45)
plt.yticks(np.arange(452000, 932000, 20000))

plt.title("Stock Price", fontsize = 30)

plt.plot(nc['date'], nc['stock_price'], marker = '^')

plt.show()
```
